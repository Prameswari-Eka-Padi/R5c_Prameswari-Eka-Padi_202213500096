```python
# nomor 3
def jumlah_hari(bulan):
  hari_per_bulan = {
      1: 31, #januari
      2: 29, #februari
      3: 31, #maret
      4: 30, #april
      5: 31, #mei
      6: 30, #juni
      7: 31, #juli
      8: 31, #agustus
      9: 30, #september
      10: 31, #oktober
      11: 30, #november
      12: 31 #desember
  }

  if bulan in hari_per_bulan:
    return hari_per_bulan[bulan]
  else:
    return None

try:
  bulan = int(input("Masukkan bulan (1-12): "))

  jumlah = jumlah_hari(bulan)

  if jumlah:
    print(f"jumlah hari: {jumlah}")
  else:
    print("bulan yang diinputkan tidak valid")
except ValueError:
  print("input tidak valid, masukkan nomor bulan berupa angka")
```

```python
# nomor 4
def cek_segitiga():
  try:
    sisi1 = float(input("Masukkan sisi 1: "))
    sisi2 = float(input("Masukkan sisi 2: "))
    sisi3 = float(input("Masukkan sisi 3: "))

    if sisi1 == sisi2 == sisi3:
      print("3 sisi sama")
    elif sisi1 == sisi2 or sisi1 == sisi3 or sisi2 == sisi3:
      print("2 sisi sama")
    else:
      print("Tidak ada yang sama")

  except ValueError:
    print("input tidak valid, Masukkan angka yang benar")

cek_segitiga()
```
